```python

def Get_Info(lat,long):
    coord_API_endpoint = "http://api.openweathermap.org/data/2.5/weather?"
    lat_long = "lat=" + str(lat)+ "&lon=" + str(long)
    join_key = "&appid=" + "7cb21de0dfabe9c26996f1c7c25b186e"
    units = "&units=metric"
    current_coord_weather_url= coord_API_endpoint + lat_long + join_key + units
    json_data = requests.get(current_coord_weather_url).json()
    Data={
        'name':json_data['name'],
        'Latitude':lat,
        'Longitude':long,
        'temp':json_data['main']['temp'],
        'pressure':json_data['main']['pressure'],
        'humidity':json_data['main']['humidity'],
        'timezone':json_data['timezone'],
        'wind_speed':json_data['wind']['speed'],
        'weather':json_data['weather'][0]['main']  
    }
    return Data
```

```python
def GET_DATA(k,l)  :
    d_1=new_data[(new_data['weather']==k) | (new_data['weather']==l)]
    x=d_1.drop(columns=['weather']).values
    y=d_1['weather'].values
    re_x,re_y=SMOTEN().fit_resample(x,y)
    data_1_=pd.concat([pd.DataFrame(re_x),pd.DataFrame(re_y)],axis=1)
    return data_1_
```

```python
from imblearn.over_sampling import SMOTEN
d_1=GET_DATA(0,1)
d_2=GET_DATA(1,2)
d_3=GET_DATA(2,3)
d_4=GET_DATA(3,0)
Resampled_data=pd.concat([d_1,d_2,d_3,d_4],axis=0)
```
